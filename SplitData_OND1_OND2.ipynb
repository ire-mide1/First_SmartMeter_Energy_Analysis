```python
#know meter_id unique values
df['meter_id'].unique()
```

```python
#form a list of all meter_id apart from OND001 and OND002

drop_them = ['LAG001', 'ADA001', 'OSU001' , 'OND005', 'OND014',
       'OSU003', 'ADA010', 'ADA009', 'ADA008', 'ADA007', 'ADA005',
       'ADA004', 'ADA003', 'ADA002', 'OND008', 'OND006',
       'OND016', 'OYO003', 'OYO001', 'OND007', 'OSU002', 'OND012',
       'OND018', 'OND010', 'OYO002', 'OSN001', 'LAG002']
```

```python
#remove all meter_id apart from OND001 and OND002

for item in drop_them:
    df = df[df.meter_id != item]
```

```python
df
```

```python
#drop nan from the categorical meter_id column

df.dropna(subset = ["meter_id"], inplace=True)
```

```python
df
```

```python
#check the new unique_id now in the dataset

df['meter_id'].unique()
```

```python
#check info for null values..real power and apparent power have null values that are gotten by calculation

df.info()
```

```python
df.head()
```

```python
#calculate apparent power

df['apparent_power'] = df['current']*df['voltage']
```

```python
df.head()
```

```python
#calculate real power

df['real_power'] = df['power_factor']/ df['apparent_power']
```

```python
df
```

```python
df.info()

#it can be observed that real_power still has nan values
```

```python
df.describe().transpose()
```

```python
#fill nan values with zero to make sure real_power is not null

df.fillna(value= 0.0000, inplace= True)
```
