```python
baseurl = 'https://sergfuta.herokuapp.com/energy-data/'
```

```python
#function to get url
def main_request(baseurl, x ):
  r = requests.get(baseurl + f'?page={x}')
  return r.json()
```

```python
#data = main_request(baseurl, 222) #call function to check
```

```python
def parse_json(response):
  charlist= [ ]
  for item in response ['results']:
    char= {
        'id': item['id'], 
        'meter_id' : item['meter_id'],
        'timestamp': item['timestamp'],
           'current': item['current'],
           'voltage': item['voltage'],
           'frequency': item['frequency'],
           'power_factor': item['power_factor'],
           'energy': item['energy'],
           'real_power': item['real_power'],
           'apparent_power': item['apparent_power']
    }

    charlist.append(char)
  return charlist    
```

```python
parse_json(data)
```

```python
mainlist= []

for page_num in range(1,708):
  page_num= page_num
  main_request(baseurl, page_num )
  mainlist.extend(parse_json(main_request(baseurl, page_num)))

print(mainlist)
```

```python
print(len(mainlist))
```

```python
import pandas as pd
df = pd.DataFrame(mainlist)
print(df.head(), df.tail())
```
